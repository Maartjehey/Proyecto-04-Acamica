# 1. 

¿Por qué es importante la clasificación de texto?
Se estima que alrededor del 80% de toda la información no está estructurada, siendo el texto uno de los tipos más comunes de datos no estructurados. Debido a la naturaleza desordenada del texto, analizar, comprender, organizar y clasificar los datos de texto es difícil y requiere mucho tiempo, por lo que la mayoría de las empresas no lo utilizan en todo su potencial.

Aquí es donde entra la clasificación de texto con aprendizaje automático. Al usar clasificadores de texto, las empresas pueden estructurar automáticamente todo tipo de texto relevante, desde correos electrónicos, documentos legales, redes sociales, chatbots, encuestas y más de una manera rápida y rentable. Esto permite a las empresas ahorrar tiempo al analizar datos de texto, automatizar procesos comerciales y tomar decisiones comerciales basadas en datos.

¿Por qué utilizar la clasificación de texto de aprendizaje automático? Algunas de las principales razones:

Escalabilidad

Analizar y organizar manualmente es lento y mucho menos preciso. El aprendizaje automático puede analizar automáticamente millones de encuestas, comentarios, correos electrónicos, etc., a una fracción del costo, a menudo en solo unos minutos. Las herramientas de clasificación de texto son escalables para cualquier necesidad empresarial, grande o pequeña.

Análisis en tiempo real

Hay situaciones críticas que las empresas deben identificar lo antes posible y tomar medidas inmediatas (por ejemplo, crisis de relaciones públicas en las redes sociales). La clasificación de texto de aprendizaje automático puede seguir las menciones de su marca constantemente y en tiempo real, por lo que identificará información crítica y podrá tomar medidas de inmediato.

Criterios consistentes

Los anotadores humanos cometen errores al clasificar datos de texto debido a distracciones, fatiga y aburrimiento, y la subjetividad humana crea criterios inconsistentes. El aprendizaje automático, por otro lado, aplica la misma lente y criterios a todos los datos y resultados. Una vez que un modelo de clasificación de texto está correctamente entrenado, funciona con una precisión insuperable.

Fuente: https://monkeylearn.com/text-classification/

Eliminar URL's and HTML

```python
import re

def remove_URL(text):
    url = re.compile(r"https?://\$+|www\.\$+")
    return url.sub(r"", text)

def remove_html(text):
    html= re.compile(r"<.*?>")
    return html.sub(r"", text)
```

Eliminar punctuacion

```python
import string

def remove_punct(text):
    table = str.maketrans("", "", string.punctuation)
    return text.translate(table)
```

```python
train["review_body"] = train.review_body.map(lambda x: remove_URL(x))
train["review_body"] = train.review_body.map(lambda x: remove_html(x))
train["review_body"] = train.review_body.map(lambda x: remove_punct(x))
```

```python
test["review_body"] = test.review_body.map(lambda x: remove_URL(x))
test["review_body"] = test.review_body.map(lambda x: remove_html(x))
test["review_body"] = test.review_body.map(lambda x: remove_punct(x))
```

```python
from nltk.corpus import stopwords

stop = set(stopwords.words('spanish'))

def remove_stopwords(text):
    text = (word.lower() for word in text.split() if word.lower() not in stop)
    
    return " ".join(text)
```

Ahora vamos a preparar el text.

```python
from collections import Counter

def counter_word(text): 
    count = Counter()
    for i in text.values:
        for word in i.split():
            count[word] += 1
    return count
```

```python
review_body = train.review_body

counter = counter_word(review_body)
```

Vamos a deifinir el numero de palabras máximo. Esto es muy importante para las secuencias porque Tensorflow quiere el mismo sequence lenght para todos. Depende del texto pero podemos ponerlo bastante alto. Creo que para este proyecto 20 está bien. Pero normalmente vemos el largo de cada texto y no basamos en esto.

```python
num_words = len(counter)

max_length = 20
```

Hacemos el train / test split para Keras:

```python
text = train["review_body"]
target = train["product_category"]

train_size = int(train.shape[0] * 0.8)

train_sentences = text[:train_size]
train_labels = target[:train_size]

test_sentences = text[train_size:]
test_labels = target[train_size:]
```

```python
from keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=num_words)
tokenizer.fit_on_texts(train_sentences)
```

Vamos a darle números a cada palabra en nuestro text:

```python
word_index = tokenizer.word_index
```

```python
word_index
```

Los llevamos a secuencias:

```python
train_sequences = tokenizer.texts_to_sequences(train_sentences)
```

```python
train_sequences[0]
```

Vamos a hacer el "padding" para hacer los sequences de mismo length:

```python
from keras.preprocessing.sequence import pad_sequences

train_padded = pad_sequences(train_sequences, maxlen=max_length, padding = "post", truncating = "post")
```

```python
train_padded[0]
```

Ya lo tenemos en un array. Lo que hizo el padding es que en vez de poner los 9 palabras al nuestro array para estancia [0] que vemos en train sequences, puso 20 números en el array (el max_length que le dimos antes). Puso los 7 números y además muchos zeros para hacer todos los sequences el mismo tamaño.

```python
test_sequences = tokenizer.texts_to_sequences(test_sentences)

test_padded = pad_sequences(test_sequences, maxlen=max_length, padding = "post", truncating = "post")
```

Hacemos al reves:

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
def decode(text):
    return " ".join([reverse_word_index.get(i, "?") for i in text])
```

```python
print(f"Shape of train {train_padded.shape}")
print(f"Shape of test{test_padded.shape}")
```

```python
train_padded = np.asarray(train_padded, dtype=np.float64, order='C')
```

```python
train_labels = np.asarray(train_labels, dtype=np.float64, order='C')
```

```python
test_padded = np.asarray(test_padded, dtype=np.float64, order='C')
```

```python
test_labels = np.asarray(test_labels, dtype=np.float64, order='C')
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
from keras.initializers import Constant
from keras.optimizers import SGD

model = Sequential()

model.add(Embedding(num_words, 32, input_length = max_length)) # Embedding es como vectorizar pero con valores reales. 
model.add(LSTM(64, dropout = 0.1))
model.add(Dense(1, activation = "sigmoid"))

optimizer = SGD(learning_rate=0.01, momentum=0.9)

model.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])

model.summary()
```

```python
history = model.fit(train_padded, train_labels, epochs=20, validation_data =(test_padded, test_labels),)
```
