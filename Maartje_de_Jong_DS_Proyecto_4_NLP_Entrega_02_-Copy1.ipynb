# 1. 

¿Por qué es importante la clasificación de texto?
Se estima que alrededor del 80% de toda la información no está estructurada, siendo el texto uno de los tipos más comunes de datos no estructurados. Debido a la naturaleza desordenada del texto, analizar, comprender, organizar y clasificar los datos de texto es difícil y requiere mucho tiempo, por lo que la mayoría de las empresas no lo utilizan en todo su potencial.

Aquí es donde entra la clasificación de texto con aprendizaje automático. Al usar clasificadores de texto, las empresas pueden estructurar automáticamente todo tipo de texto relevante, desde correos electrónicos, documentos legales, redes sociales, chatbots, encuestas y más de una manera rápida y rentable. Esto permite a las empresas ahorrar tiempo al analizar datos de texto, automatizar procesos comerciales y tomar decisiones comerciales basadas en datos.

¿Por qué utilizar la clasificación de texto de aprendizaje automático? Algunas de las principales razones:

Escalabilidad

Analizar y organizar manualmente es lento y mucho menos preciso. El aprendizaje automático puede analizar automáticamente millones de encuestas, comentarios, correos electrónicos, etc., a una fracción del costo, a menudo en solo unos minutos. Las herramientas de clasificación de texto son escalables para cualquier necesidad empresarial, grande o pequeña.

Análisis en tiempo real

Hay situaciones críticas que las empresas deben identificar lo antes posible y tomar medidas inmediatas (por ejemplo, crisis de relaciones públicas en las redes sociales). La clasificación de texto de aprendizaje automático puede seguir las menciones de su marca constantemente y en tiempo real, por lo que identificará información crítica y podrá tomar medidas de inmediato.

Criterios consistentes

Los anotadores humanos cometen errores al clasificar datos de texto debido a distracciones, fatiga y aburrimiento, y la subjetividad humana crea criterios inconsistentes. El aprendizaje automático, por otro lado, aplica la misma lente y criterios a todos los datos y resultados. Una vez que un modelo de clasificación de texto está correctamente entrenado, funciona con una precisión insuperable.

Fuente: https://monkeylearn.com/text-classification/

Eliminar URL's and HTML

```python
import re

def remove_URL(text):
    url = re.compile(r"https?://\$+|www\.\$+")
    return url.sub(r"", text)

def remove_html(text):
    html= re.compile(r"<.*?>")
    return html.sub(r"", text)
```

Eliminar punctuacion

```python
import string

def remove_punct(text):
    table = str.maketrans("", "", string.punctuation)
    return text.translate(table)
```

```python
train["review_body"] = train.review_body.map(lambda x: remove_URL(x))
train["review_body"] = train.review_body.map(lambda x: remove_html(x))
train["review_body"] = train.review_body.map(lambda x: remove_punct(x))
```

```python
from collections import Counter

def counter_word(text): 
    count = Counter()
    for i in text.values:
        for word in i.split():
            count[word] += 1
    return count
```

Ahora separamos el train del test. X son nuestras secuencias rellenada: todas estén en la misma longitud para el modelo. Para Y convertimos los *labels* categoricos a números. 

```python
from sklearn.model_selection import train_test_split


X = train_padded[:5000]
Y = pd.get_dummies(df_train['product_category']).values[:5000]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.10, random_state = 42)
                  
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

Explicamos el modelo: 
- La primera capa es la Embedding que utiliza 100 vectores de longitud para representar cada palabra.

- SpatialDropout1D promueve la independencia entre los mapas de caractericticás - 1D 

- La siguiente capa es la capa LSTM con 100 unidades de memoria.

- La capa de salida debe crear 30 valores de salida, uno para cada clase en nuestra columna product_category.

- La función de activación es softmax para la clasificación de clases multiples.

- Debido a que es un problema de clasificación de varias clases, categorical_crossentropy se utiliza como la función de *loss*.
La Cross-entropy es la función de pérdida por defecto que se utiliza para los problemas de clasificación multiclase.

En este caso, está pensada para su uso con la clasificación multiclass en la que los valores objetivo se encuentran en el conjunto {0, 1, 3, ..., n}, donde a cada clase se le asigna un único valor integer.

Matemáticamente, es la función de pérdida preferida en el marco de la inferencia de máxima probabilidad. Es la función de pérdida que debe evaluarse en primer lugar y sólo se cambia si se tiene una buena razón.

La Cross-entropy calculará una puntuación que resume la diferencia media entre las distribuciones de probabilidad real y predicha para todas las clases del problema. La puntuación se minimiza y un valor Cross-entropy perfecto es 0.

La Cross-entropy puede especificarse como loss function en Keras especificando 'categorical_crossentropy' al compilar el modelo.

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout, Masking, Activation, SpatialDropout1D
from keras.initializers import Constant
from keras.optimizers import SGD


model = Sequential()
model.add(Embedding(num_words, 100, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(30, activation='softmax'))


model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=["accuracy"])

model.summary()
```

El problema con las redes neuronales está en la elección del número de epochs  a utilizar.

Demasiadas epochs pueden llevar a la *overfitting* del conjunto de datos de entrenamiento, mientras que demasiado pocas pueden resultar en un modelo de *underfitting*. Early Stopping es un método que permite especificar un gran número arbitrario de épocas de entrenamiento y detener el entrenamiento una vez que el rendimiento del modelo deja de mejorar en un conjunto de datos de validación de espera.

```python
from keras.callbacks import EarlyStopping
epochs = 20
batch_size = 64

history = model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size,validation_split=0.1,callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
```

```python
accr = model.evaluate(X_test,Y_test)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```

```python
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show();
```

```python
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
plt.show();
```

Hemos creado dos lineplots. El de arriba nos enseña el cross-entropy loss a lo largo de las epochs para el conjunto de datos de entrenamiento (azul) y de prueba (naranja), y el gráfico inferior que muestra la precisión de la clasificación a lo largo de las épocas.

Vemos que tenemos un model super underfitteado. No habia tanto Loss en el Train, pero en el Test si. También el accuracy se queda en cero con el test set. Podriamos ajustar el learning rate o batch size igualar la convergencia en este caso.
